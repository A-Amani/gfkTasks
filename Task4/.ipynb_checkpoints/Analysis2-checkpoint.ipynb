```python
import pandas as pd
import numpy as np
```

```python
x = [1400, 1600, 1700, 1875, 1100, 1550, 2350, 2450, 1425, 1700] # house area in sq-feet
y = [245, 312, 279, 308, 199, 219, 405, 324, 319, 255] # price in $1000s
```

First, let's see how the data looks like:

```python
import matplotlib.pyplot as plt

np.random.seed(19680801)
plt.scatter(x, y, alpha=0.75)
plt.xlabel('square-feet size')
plt.ylabel('House price in 1K $')
plt.show()
```

Let's fit a linear model into the data and take a look at the model:


```python
from sklearn import linear_model

x = np.array(x).reshape(-1, 1)
y = np.array(y)

regr = linear_model.LinearRegression()

regr.fit(x, y)
```

Let's see how the model looks like:

Can I improve the model behaviour by adding polynomial features? 

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import cross_val_score


degrees = [2, 3, 5]

plt.figure(figsize=(15, 5))

for i in range(len(degrees)):

    ax = plt.subplot(1, len(degrees), i + 1)

    polynomial_features = PolynomialFeatures(degree=degrees[i], include_bias=False)
    linear_regression = linear_model.LinearRegression()
    pipeline = Pipeline([("polynomial_features", polynomial_features),
                         ("linear_regression", linear_regression)])
    pipeline.fit(x, y)
    scores = cross_val_score(pipeline, x, y, scoring="neg_mean_squared_error")
    
    plt.plot(xp, pipeline.predict(xp), '-r', label="Model")
    plt.scatter(x, y, edgecolor='r', label="samples")
    plt.xlabel('square-feet size')
    plt.ylabel('House price in 1K $')
    plt.xlim(1000,2500)
    plt.ylim(100,500)
    plt.legend(loc="best")
    plt.title("Degree {}\nMSE = {:.2e}(+/- {:.2e})".format(degrees[i], -scores.mean(), scores.std()))

plt.show()
```

Judging based on MSE, the model with 2nd-order polynomial features works the best!
But better check Goodness of fit through other metrics


**Goodness of fit**

Let's check the coefficient of determination R^2 of the models:


```python
degrees = [1, 2, 3, 5]

plt.figure(figsize=(14, 5))

for i in range(len(degrees)):

    ax = plt.subplot(1, len(degrees), i + 1)

    polynomial_features = PolynomialFeatures(degree=degrees[i],include_bias=False)
    linear_regression = linear_model.LinearRegression()
    pipeline = Pipeline([("polynomial_features", polynomial_features),
                         ("linear_regression", linear_regression)])
    pipeline.fit(x, y)
    scores = pipeline.score(x, y)

    plt.plot(xp, pipeline.predict(xp), '-r', label="Model")
    plt.scatter(x, y, edgecolor='r', label="samples")
    plt.xlabel('square-feet size')
    plt.ylabel('House price in 1K $')
    plt.xlim(1000,2500)
    plt.ylim(100,500)
    plt.legend(loc="best")
    plt.title("Degree {}\n R^2 = {:.2e}".format(degrees[i], scores.mean(), scores.std()))    
```

In R^2, the best possible score is 1.0. So according to this criteria, the model with 5th-order polynomials is performing the best. Eventhough the MSE of model with 2nd-order polynomial features is lower!

One problem with the R2 as a measure of model validity is that it can always be increased by adding more variables into the model, except in the unlikely event that the additional variables are exactly uncorrelated with the dependent variable in the data sample being used. 
So in case of adding up to 5th-order polynomial features into the data, the R^2 has increased as predicted!
This problem can be avoided by  using the adjusted R2.



```python
degrees = [1, 2, 3, 5]

plt.figure(figsize=(15, 5))

for i in range(len(degrees)):

    ax = plt.subplot(1, len(degrees), i + 1)

    polynomial_features = PolynomialFeatures(degree=degrees[i],include_bias=False)
    linear_regression = linear_model.LinearRegression()
    pipeline = Pipeline([("polynomial_features", polynomial_features),
                         ("linear_regression", linear_regression)])
    pipeline.fit(x, y)
    scores = pipeline.score(x, y)
    n = x.shape[0]
    p = degrees[i]
    scores = 1.-(1.-scores*scores)*(n-1.)/(n-p-1.)
    
    plt.plot(xp, pipeline.predict(xp), '-r', label="Model")
    plt.scatter(x, y, edgecolor='r', label="samples")
    plt.xlabel('square-feet size')
    plt.ylabel('House price in 1K $')
    plt.xlim(1000,2500)
    plt.ylim(100,500)
    plt.legend(loc="best")
    plt.title("Degree {}\n adjusted R^2 score = {:.2e}".format(degrees[i], scores.mean()))    
```

So according to the adjusted R^2 score, the 1st-degree polynomial feature model, fits best the data ( and not the 5th-order features, as suggested earlier by R^2!)

**F-test statistics and mutual information**
Eventhough the simple linear model is selected as the best fit, for the sake of personal curiosity (:D), I will check the F-test statistics and mutual information of different features in 5th-order polynomial feature model.
The code below plots the dependency of price against individual polynomia and normalized values of univariate F-tests statistics and mutual information.

```python
from sklearn.feature_selection import f_regression, mutual_info_regression
deg = 5

polynomial_features = PolynomialFeatures(degree=deg, include_bias=False)
xtmp = polynomial_features.fit_transform(x)

f_test, _ = f_regression(xtmp, y)
f_test   /= np.max(f_test)

mi  = mutual_info_regression(xtmp, y)
mi /= np.max(mi)

plt.figure(figsize=(15, 5))
for i in range(deg):
    plt.subplot(1, deg, i + 1)
    plt.scatter(xtmp[:, i], y, edgecolor='black', s=20)
    plt.xlabel(f'x^{i+1} features', fontsize=14)
    if i == 0:
        plt.ylabel("$y$", fontsize=14)
    plt.title("F-test={:.2f}, MI={:.2f}".format(f_test[i], mi[i]), fontsize=16)
plt.show()
```

As F-test captures only linear dependency, it rates x as the most discriminative feature. On the other hand, mutual information can capture any kind of dependency between variables and it rates x^3 as the most discriminative feature

**Let's go back to the winner model. Let's creat a formula string for using in the statsmodels.OLS() and use it to extract more statistical insight of the model**


```python
import statsmodels.formula.api as sm
df  = pd.DataFrame(x, columns=['x'])
deg = 1
for i in range(2,deg+1):
    df[f'x^{i}'] = df['x'].pow(i)
df['price'] = y

formula_str = df.columns[-1]+' ~ ' + '+'.join(df.columns[:-1])
formula_str
```

```python
model=sm.ols(formula=formula_str, data=df)
fitted = model.fit()
print(fitted.summary())
```

**Let's go back a little and check out p-values and statistical significance of the features**

```python
df_result=pd.DataFrame()
df_result['pvalues']=fitted.pvalues[1:]
df_result['Features']=df.columns[:-1]
df_result.set_index('Features',inplace=True)

def yes_no(b):
    if b<0.05:
        return 'Yes'
    else:
        return 'No'
    
df_result['Statistically significant?']= df_result['pvalues'].apply(yes_no)
df_result
```

We know a data set provides statistical significance when the p-value is sufficiently small.
When the p-value is large, then the results in the data are explainable by chance alone, and the data are deemed consistent with (while not proving) the null hypothesis.
In this case the feature shows to be statistically significant! :)

**Residuals vs. predicting variables plots**


```python
plt.figure(figsize=(4, 5))

for i, c in enumerate(df.columns[:-1]):
    plt.subplot(1, len(df.columns)-1, i + 1)
    plt.title("{} vs. \nModel residuals".format(c),fontsize=16)
    plt.scatter(x=df[c],y=fitted.resid,color='blue',edgecolor='k')
    plt.grid(True)
    xmin=min(df[c])
    xmax = max(df[c])
    plt.hlines(y=0,xmin=xmin*0.9,xmax=xmax*1.1,color='red',linestyle='--',lw=3)
    plt.xlabel(c,fontsize=14)
    plt.ylabel('Residuals',fontsize=14)
    
plt.show()
```

The distribution seem random around the 0 axis 
So, overall the assumptions of linearity and independence seem to hold, but again we only have 10 data samples! 


```python
from statsmodels.graphics.gofplots import qqplot


plt.figure(figsize=(8,5))
fig=qqplot(fitted.resid_pearson,line='45',fit='True')
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.xlabel("Theoretical quantiles",fontsize=15)
plt.ylabel("Sample quantiles",fontsize=15)
plt.title("Q-Q plot of normalized residuals",fontsize=18)
plt.grid(True)
plt.show()
```

The Q-Q plot shows that the normality assumption is satisfied pretty good :)


**We can check for outliers in residuals by e.g. Cook's distance:**


```python
from statsmodels.stats.outliers_influence import OLSInfluence as influence
inf=influence(fitted)
(c, p) = inf.cooks_distance
plt.figure(figsize=(8,5))
plt.title("Cook's distance plot for the residuals",fontsize=16)
plt.stem(np.arange(len(c)), c, markerfmt=",")
plt.grid(True)
plt.show()
```
